# Aprendizaje

XGBoost no entiende el tiempo por si mismo, por eso es necesario darle "lags" que es practicamente decirle que el pasado importa para las predicciones siguientes que va a hacer, en la columna lag_1 por ejemplo para un producto se mostrara las ventas de hace la semana pasada, lag_2 mostrara las ventas de hace 2 semanas

asi el modelo entiende que las predicciones se hacen en base a las semanas anteriores y en como estos patrones se comportan

tambien a partir de los lags se puede obtener una media y un trend, pero primero unicamente usar los lags 

tambien a la hora de hacer tuning del modelo sera importante validar cuantos lag dejar, que seria cuantas semanas en el pasado observar para la prediccion de cada producto.
